```python
from matplotlib import pyplot as plt
def gradient_descent(x,y):
    m, b = 0, 0
    n = len(x)
    lr = 0.001
    plt.scatter(x, y)
    for i in range(1000):
        y_pred = m*x+b  
        plt.plot(x, y_pred)
        mse = sum((y-y_pred)**2)/n
        print(f"m: {m}, b: {b}, mse: {mse}, i: {i}")
        dm = -2*sum(x*(y-y_pred))/n
        m = m-dm*lr
        db = -2*sum(y-y_pred)/n
        b = b-db*lr      
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
%matplotlib inline
gradient_descent(x,y)
```
